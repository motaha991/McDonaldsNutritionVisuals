# McDonald's Nutrition Facts Analysis

## Overview
This project involves the analysis of McDonaldâ€™s nutrition data. The dataset contains details about various menu items, including their nutritional values such as calories, fats, carbohydrates, and more. The analysis is conducted using Python libraries like `pandas`, `matplotlib`, and `seaborn`.

```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

**Loading the Data**: The dataset is loaded using `pandas` from a CSV file.

```python
df = pd.read_csv('/content/menu.csv')
df.head()
```

**Summary Statistics**: Descriptive statistics of the dataset are calculated using the `describe()` method.

```python
df.describe()
```

**Visualization**: Various visualizations are generated to explore the nutritional data


- **Calories Distribution**: A histogram shows the distribution of calories across menu items.

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Calories'], kde=True)
plt.title('Distribution of Calories')
plt.xlabel('Calories')
plt.ylabel('Frequency')
plt.show()
```

- **Correlation Heatmap**: A heatmap visualizes the correlation between different nutritional values.

```python
# Calculate the correlation matrix for calories, protein, carbohydrates, and sugars
nutritional_factors = df[['Calories', 'Protein', 'Carbohydrates', 'Sugars', 'Total Fat']]
correlation_matrix = nutritional_factors.corr()

plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Between Calories, Protein, Carbohydrates, Sugars, and Fat')
plt.show()
```

- **Calories Distribution per Category:** A boxplot visualizes the range and distribution of calories for each category, helping identify categories with wide calorie variation or consistently high/low calorie counts.


```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Category', y='Calories', data=df)
plt.xticks(rotation=90)
plt.title('Calories Distribution per Category')
plt.show()
```

## Key Findings
- **Calorie Ranges**: Most items fall between 200 to 500 calories, with a few high-calorie items exceeding 1000 calories.
- **Correlation Analysis**: Calories show a high correlation with fat content, while carbohydrates correlate with sugars.
- **Categories with Higher Caloric Spread(BoxPlot)**: Categories such as _Breakfast_ or _Burgers_ have **larger boxes** and **longer whiskers**, indicating a wide range of calorie values. This suggests that some items in these categories are much higher in calories than others.

- **Categories with Lower Caloric Spread(BoxPlot)**: Categories like _Beverages_ or _Salads_ have **narrower boxes** and **shorter whiskers**, indicating that most items in these categories have similar calorie counts.

- **Outliers(BoxPlot)**: Some items have significantly more calories than the rest in a category (e.g., a particularly large or loaded burger), these appear as **outlier points** above the whiskers. Similarly, low-calorie items appear as **outlier points** below the whiskers.


## Conclusion
The analysis helps to better understand the nutritional content of McDonald's menu items. By visualizing the data and summarizing key metrics, we can identify items that are high in calories, fat, and other nutritional values.
